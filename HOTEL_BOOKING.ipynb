## 3. ***Data Wrangling***

### Data Wrangling Code

```python
# Write your code to make your dataset analysis ready.
import pandas as pd
import numpy as np

dataset = pd.read_excel('/content/drive/MyDrive/newdataset/Hotel Bookings (1).xlsx')

print("Initial shape:", dataset.shape)
print("Missing values:\n", dataset.isnull().sum())

threshold = 0.8
dataset = dataset.loc[:, dataset.isnull().mean() < threshold]

columns_to_check = ['children', 'country', 'agent', 'company']
for col in columns_to_check:
    if col in dataset.columns:
        if dataset[col].dtype == 'object':
            dataset[col] = dataset[col].fillna('Unknown')
        else:
            dataset[col] = dataset[col].fillna(0)

if 'reservation_status_date' in dataset.columns:
    dataset['reservation_status_date'] = pd.to_datetime(dataset['reservation_status_date'])



print("Cleaned shape:", dataset.shape)
print("Any remaining missing values?\n", dataset.isnull().sum().sum())

```

```python
# Chart - 2 visualization code
import pandas as pd
import matplotlib.pyplot as plt

guest_comp = dataset.groupby('hotel')[['adults', 'children', 'babies']].mean().reset_index()

labels = guest_comp['hotel']
adults = guest_comp['adults']
children = guest_comp['children']
babies = guest_comp['babies']

plt.figure(figsize=(8, 6))
plt.bar(labels, adults, label='Adults', color='skyblue')
plt.bar(labels, children, bottom=adults, label='Children', color='lightgreen')
plt.bar(labels, babies, bottom=adults+children, label='Babies', color='salmon')

plt.title('Average Guest Composition per Booking by Hotel Type')
plt.xlabel('Hotel Type')
plt.ylabel('Average Number of Guests')
plt.legend()
plt.tight_layout()
plt.show()
```

#### Chart - 14 - Correlation Heatmap

```python
# Correlation Heatmap visualization code
import matplotlib.pyplot as plt
import seaborn as sns

numeric_data = dataset.select_dtypes(include=['int64', 'float64'])
corr_matrix = numeric_data.corr()

plt.figure(figsize=(20, 24))
sns.heatmap(corr_matrix, annot=True, cmap='coolwarm', fmt=".2f", square=True)
plt.title('Correlation Heatmap of Numerical Features')
plt.tight_layout()
plt.show()
```

##### 1. Why did you pick the specific chart?

Answer Here.

I chose the **correlation heatmap** because it is the most effective way to visualize the **strength and direction of relationships between numerical variables** in a dataset. This chart makes it easy to identify which features are strongly correlated—positively or negatively—which is critical for **feature selection, multicollinearity detection, and predictive modeling**. By displaying all pairwise correlations in a single, color-coded matrix, it helps uncover hidden patterns, such as whether variables like **lead time**, **ADR**, or **number of special requests** influence **cancellations** or other business outcomes.

##### 2. What is/are the insight(s) found from the chart?

*Answer* Here

The correlation heatmap reveals several key insights:

* There is a **moderate positive correlation between lead time and cancellation** (`is_canceled`), indicating that bookings made further in advance are more likely to be canceled.
* **Special requests have a negative correlation with cancellations**, suggesting that guests who make requests are more likely to follow through with their stay.
* Variables like **total nights stayed, number of adults, and children** show little to no correlation with cancellations, meaning these features may have limited predictive power for that outcome.
* Some internal relationships are also evident—for example, **total nights** (sum of week and weekend nights) strongly correlates with both components, which is expected.

These patterns help in understanding **which variables drive key outcomes**, and guide better feature selection for modeling or operational strategy.
